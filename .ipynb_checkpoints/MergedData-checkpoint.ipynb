```python
# Dependency
import pandas as pd
```

```python
# Load in CSVs
jan_data = "Resources/JC-202201-citibike-tripdata.csv"
feb_data = "Resources/JC-202202-citibike-tripdata.csv"
mar_data = "Resources/JC-202203-citibike-tripdata.csv"
```

```python
# Read CSVs
jan_pd = pd.read_csv(jan_data)
feb_pd = pd.read_csv(feb_data)
mar_pd = pd.read_csv(mar_data)
```

```python
# Concatenate all 3 dataframes into one
nycbike_q1 = pd.concat([jan_pd,feb_pd,mar_pd])
```

```python
# Print dataframe
nycbike_q1.head()
```

```python
# Export file as a CSV
nycbike_q1.to_csv("Resources/MergedData.csv", index=False, header=True)
```
